<h2><strong>Meaningful Insights on Child Mortality In Africa</strong></h2>

```python
plt.style.use('ggplot')
plt.figure(figsize=(15,15))
plt.bar(merged_df1['Year'],merged_df1['Under-fifteen mortality rate'], color='chocolate')
plt.title('Under-fifteen Mortality Rate Over Years',fontweight='bold', fontsize=17)
plt.xlabel('Year',fontweight='bold', fontsize=14)
plt.ylabel('Under-fifteen Mortality Rate',fontweight='bold', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(13,13))
plt.style.use('bmh')
plt.barh(merged_df1['Year'],merged_df1['Deaths - Sex: all - Age: 0 - Variant: estimates'],color='darkslategrey')
plt.title('Estimated Deaths Over Years', fontweight='bold', fontsize=14)
plt.xlabel('Estimated Deaths (Age 0, All Sexes)', fontweight='bold', fontsize=12)
plt.ylabel('Year', fontweight='bold', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(15,15))
plt.style.use('Solarize_Light2')
sns.barplot(x='Year',y='Observation value - Indicator: Under-five mortality rate - Sex: Total - Wealth quintile: Total - Unit of measure: Deaths per 100 live births',data=merged_df2)
plt.xticks(rotation=90)
plt.ylabel('Deaths per 100 live births')
plt.title('Deaths per 100 live births over the years',fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(15,15))
plt.style.use('bmh')
sns.regplot(x='Births attended by skilled health staff (% of total)',y='Estimated maternal deaths',data=merged_df3,color='firebrick')
plt.title('Relationship between Estimated Maternal Deaths and Births Attended By Health Staff',fontweight='bold',fontsize=13)
plt.show()
```

```python
plt.figure(figsize=(15,15))
plt.style.use('bmh')
ax = sns.stripplot(x='Dim2',y='FactValueNumeric',data=df_4,color='mediumslateblue')
plt.title('Fact Values For Each Cause Of Death',fontsize=15,fontweight='bold')
plt.xlabel('Cause of Death',fontweight='bold',fontsize=15)
plt.ylabel('Fact Values',fontweight='bold',fontsize=15)
plt.xticks(rotation=90,fontweight='bold',fontsize=15)
ax.set_facecolor('lightblue')
plt.show()
```

<h1><strong>Actionable Recommendations To Stop Infant Mortality</strong></h1>

<h2>1. Improving Vaccination Coverage</h2>
        <h3>A. Community-Based Interventions</h3>
        <ul>
            <li><strong>Mobile Vaccination Clinics:</strong> Deploy mobile teams to underserved areas.</li>
            <li><strong>Home-Based Vaccination:</strong> Provide door-to-door vaccination services.</li>
            <li><strong>School-Based Programs:</strong> Partner with schools to reach children efficiently.</li>
        </ul>

<h3>B. Public Awareness Campaigns</h3>
        <ul>
            <li>Launch campaigns to educate families about the importance of vaccines.</li>
            <li>Combat misinformation through workshops and social media.</li>
            <li>Engage <strong>trusted local leaders</strong> to endorse vaccination efforts.</li>
        </ul>

        

<h3>C. Incentivize Vaccination</h3>
        <ul>
            <li>Offer small incentives like food supplies or health checkups for completing vaccinations.</li>
            <li>Use SMS reminders to keep track of vaccination appointments.</li>
        </ul>

 <h2 class="section-title">2. Enhancing Healthcare Access</h2>
        <h3>A. Expand Healthcare Infrastructure</h3>
        <ul>
            <li>Build or renovate clinics in underserved areas.</li>
            <li>Equip facilities with essential tools, medicine, and staff.</li>
        </ul>

        

<h3>B. Improve Transportation and Outreach</h3>
        <ul>
            <li>Provide <strong>ambulance services</strong> or transport vouchers for families.</li>
            <li>Partner with NGOs to organize community health events.</li>
        </ul>

        

<h3>C. Train and Deploy Local Health Workers</h3>
        <ul>
            <li>Train community health workers to provide basic maternal and child health services.</li>
            <li>Implement task-shifting programs for nurses and midwives.</li>
        </ul>

        


<h3>D. Strengthen Digital Health Systems</h3>
        <ul>
            <li>Use <strong>telemedicine</strong> to provide remote healthcare consultations.</li>
            <li>Develop mobile health apps to track patient data.</li>
        </ul>

<h2 class="section-title">3. Providing Maternal Support</h2>
        <h3>A. Improve Antenatal and Postnatal Care</h3>
        <ul>
            <li>Offer free or subsidized prenatal checkups for mothers.</li>
            <li>Provide postnatal follow-ups to monitor maternal and infant health.</li>
        </ul>

       

 <h3>B. Nutrition Programs</h3>
        <ul>
            <li>Introduce maternal and infant nutrition programs.</li>
            <li>Provide essential supplements like iron, folic acid, and Vitamin A.</li>
        </ul>

        

<h3>C. Promote Maternal Education</h3>
        <ul>
            <li>Offer workshops on safe delivery, breastfeeding, and newborn care.</li>
            <li>Educate mothers to recognize signs of illness in infants.</li>
        </ul>

        

<h3>D. Support Mental and Financial Wellbeing</h3>
        <ul>
            <li>Set up <strong>maternal support groups</strong> to create a network of care.</li>
            <li>Provide financial support, like cash transfers or microloans, during pregnancy and after birth.</li>
        </ul>
